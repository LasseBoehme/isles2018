```python
from keras.models import load_model
import keras
from keras import backend as K
from scipy.spatial.distance import dice
import numpy as np
import re
import os
import glob
import nibabel as nib
from skimage.transform import resize
from skimage import img_as_bool
import matplotlib.pyplot as plt
```

```python
from isles_data import load_imgs


imgs, gts, add = load_imgs("/home/lasse/Dokumente/isles18/TRAINING", 128)
```

```python
from task import get_kfold_indices
from loss_functions import dice_coef, dice_coef_loss

model_path = "/home/lasse/Dokumente/isles18/gctest/cloud_models/model_depth3_k{}.h5"

# die selben Splits erzeugen wie zum Trainieren


all_dcs = []
for i, (train_indices, test_indices) in enumerate(get_kfold_indices(imgs, 4)):
    print(test_indices)
    
    model = load_model(model_path.format(i), custom_objects={"dice_coef" : dice_coef, "dice_coef_loss" : dice_coef_loss})
    dcs = []
    
    for test_index in test_indices:
        img = model.predict(imgs[test_index], verbose=0).squeeze()
        
        dc = 1.0 - dice(img.flatten(), gts[test_index].flatten())
        dcs.append(dc)
        all_dcs.append(dc)
        print("{}({}): {:0.2f} {}".format(add[test_index]["case"], add[test_index]["id"], dc, img.max()))
    
    print("DC_k{} = {:0.2f}".format(i, np.mean(dcs)))
    print("")
    
print("")
print("Alles DCs = {:0.2f}".format(np.mean(all_dcs)))
```
